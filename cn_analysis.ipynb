### For all notes categorized as needs more ratings, how long have they been stuck in that rating? And for how long?

```python
df = pd.read_csv("noteStatusHistory.tsv", sep="\t", low_memory=False)
df
```

```python
df.info()
```

```python
#Converting creation time of note (in milliseconds) to datetime format
```

```python
df['created_time'] = pd.to_datetime(df['createdAtMillis'], unit='ms').dt.floor('s')
```

```python
#Filtering for all notes whose current status is "NEEDS_MORE_RATINGS"
```

```python
new_df = df[df['currentStatus'] == 'NEEDS_MORE_RATINGS']
new_df
```

```python
#Assigning current day: February 11, 2025
new_df['current_day'] = pd.to_datetime('2025-02-11 00:00:00')
```

```python
#Subtracting current_day from all created_time entries to see how long the notes have been in the NMR status since they were first created
new_df.loc[:, 'difference'] = new_df['current_day'] - new_df['created_time']
```

```python
#Filtering all notes that have been stuck in NMR status for 30+ days
new_df = new_df[new_df['difference'] > '30 days']
new_df
```

```python
#Removing the time in hours and only keeping the days
import re
new_df['difference'] = new_df['difference'].astype(str)
new_df['difference'] = new_df['difference'].str.extract(r'(\d+) days')[0]
new_df
```

```python
new_df = new_df.copy()
new_df['difference'] = new_df['difference'].astype(float)

#making bins to put each note in a bin

bins = [0, 30, 180, 365, 730, float('inf')]  # Ranges
labels = ['one month', '1-6 months', '6 months-1year', '1-2 years', '2years+']  # Labels

# Categorize days into bins
new_df.loc[:, 'day_range'] = pd.cut(new_df['difference'], bins=bins, labels=labels, right=True).astype(pd.CategoricalDtype(categories=labels, ordered=True))


# Group by day_range and count each noteId to see how many notes fall in that range
grouped_counts = new_df.groupby('day_range', observed=False)['noteId'].count().reset_index()

# Renaming columns
grouped_counts.rename(columns={'noteId': 'note_count'}, inplace=True)
```

```python
grouped_counts
```

```python
grouped_counts.to_csv("nmr_grouped_by_days.csv")
```

### How were misleading posts categorized by contributors?

```python
#loading new dataframe
df = pd.read_csv("notes-00000.tsv", sep="\t", low_memory=False)  

#converting note creation time in milliseconds to datetime
df['datetime'] = pd.to_datetime(df['createdAtMillis'], unit='ms').dt.floor('s')

# Setting which columns to count
columns_to_count = [
    'misleadingOther', 
    'misleadingFactualError', 
    'misleadingManipulatedMedia', 
    'misleadingOutdatedInformation', 
    'misleadingMissingImportantContext', 
    'misleadingUnverifiedClaimAsFact', 
    'misleadingSatire'
]

# only counting rows where value for columns is 1, grouping by year and adding them up
yearly_counts = (
    df[columns_to_count].eq(1)    
    .groupby(df['datetime'].dt.year)  
    .sum()                       
)


yearly_counts = yearly_counts.reset_index()

# Display the result
yearly_counts
```

```python
yearly_counts.to_csv("year-wise-claim-categories.csv")
```
